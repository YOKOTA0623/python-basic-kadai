【課題５】Matplotlibで、可視化した複数のグラフを1つの描画キャンパス上で同時に表示しよう

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2,3,tight_layout=True )

#棒グラフ作成
x0 = [0,2]
y0 = [1.5,2,-5,2]
axes[0,0].plot(y0)

#sin関数のプロット作成
x0 = np.linspace(0, 10, 100)
y0 = 2 + 2 * np.sin(2 * x0)
axes[0,1].plot(x0, y0, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

#ヒストグラム作成
x0 = np.random .normal(15,5,2000)
plt.xlim(-10, 60)
axes[0,2].hist(x0, bins=10)

# 散布図作成
np.random.seed(3)
x0 = 4 + np.random.normal(0, 2, 60)
y1 = 4 + np.random.normal(0, 2.5, len(x0))
axes[1, 0].scatter(x0, y1)


plt.show()





```
