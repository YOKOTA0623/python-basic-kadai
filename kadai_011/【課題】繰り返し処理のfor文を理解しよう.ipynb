```python
for i in range(1,11):
  print(i)
```

for               
繰り返す回数があらかじめわかっている場合。または配列や辞書（ 連想配列） に対して繰り返し処理を行いたい                                                                                                          
  ・1～10までの数字を順番に表示する                                            
                ・配列の中に入っている都道府県の名前を順番に取り出し、表示する   

```python
for i in range(3,12,2):
  print(i)
```

```python
for i in range(1,11):
  print(i)

  if i == 5:
    break
```

```python
#ランダムな整数を利用するためrandom をインポート
import random

for i in range(1,11):
  num = random.randint(1,20)

  print(f"{i}回目の結果は{num}です。")

  if num == 20:
    print("20がでたので繰り返し処理を強制終了します")
    break
```

```python
for i in range(1,11):
 if i % 2 == 1 :
  continue

 print(i)
```

```python
import random

num = random.randint(0,4)

print(f"最初の値{num}です")

while num != 0:
  num = random.randint(0,4)

  print (f"現在の値は{num}です")

```

```python
import random

sum=0

while sum < 20:
  num = random.randint(1,10)
  print(f"{num}がでました")

  if num % 2 == 0:
    print("偶数なので加算しません")
    continue

  sum = sum +num
  print(f"現在の合計は{sum}です")
```

```python
# 11.3 繰り返し処理の書き方（配列の場合）

users_nemes = ["侍太郎","侍一郎","侍二郎","侍三郎","侍四郎"]

for user_name in users_nemes:
  print(user_name)
```

```python
personal_data = {"name":"侍太郎","age":36,"gender":"男性"}

# 連想配列 personal_dataのキーと値を１つずつ順番に出力する
for key,value in personal_data.items():
  print(f"{key}は{value}です")

#連想配列
for value in personal_data.values():
  print(value)
```

```python
user_nemes = ["侍太郎","侍一郎","侍二郎","侍三郎","侍四郎"]

#検索したいユーザー名を代入する変数
target = "侍二郎"

for user_name in user_nemes:
  print(user_name)

  #変数user_name と変数targetの値が一致すれば、break文で繰り返し処理を強制終了する

  if(user_name == target):
    print(f"{target}さんがみつかったので、繰り返し処理を強制終了します")
    break
```

```python
score = {"国語":80,"数学":55,"理科":70,"社会":85,"英語":69}

print("合格した科目は以下の通りです")

for key, value in score.items():
  if(value <70):
    continue
  print(f"{key}:{value}点")
```
